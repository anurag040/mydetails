```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
import seaborn as sns
import unittest

# Load Transaction Data
file_path = "Combined_File.xlsx"  # Update with your actual file path
df = pd.read_excel(file_path)

# Load Market Announcements Data
announcements_path = "market_announcements.csv"  # Update with your announcements file path
announcements = pd.read_csv(announcements_path)

# Rename columns for consistency
df.rename(columns={
    "Trade date": "Transaction Date",
    "Transaction category": "Transaction Type",
    "Trade amount (in millions, USD)": "Transaction Volume",
    "Price": "Transaction Price",
    "Total amount transferred (in millions, USD)": "Total Amount Transferred",
    "Counterparty": "Trader_ID",
    "Security description": "Security Description",
    "CUSIP": "Security ID"
}, inplace=True)

# Convert Date columns to datetime format, handling different formats
df["Transaction Date"] = pd.to_datetime(df["Transaction Date"], errors='coerce')
announcements["Announcement Date"] = pd.to_datetime(announcements["Announcement Date"], errors='coerce', dayfirst=True)

# Merge Transactions with Announcements
df = df.merge(announcements, left_on="Transaction Date", right_on="Announcement Date", how="left")

# Feature Engineering
df["Rolling_Volume_Mean"] = df["Transaction Volume"].rolling(window=30).mean()
df["Rolling_Volume_STD"] = df["Transaction Volume"].rolling(window=30).std()
df["Rolling_Price_Mean"] = df["Transaction Price"].rolling(window=30).mean()
df["Rolling_Price_STD"] = df["Transaction Price"].rolling(window=30).std()

# Assign Impact Score based on Sentiment
df["Impact_Score"] = df["Sentiment"].map({"positive": 1, "negative": -1, "neutral": 0})

# Flag Transactions Before Announcements
df["Trade_Before_Announcement"] = np.where(df["Transaction Date"] < df["Announcement Date"], 1, 0)

# Detecting Unusual Trading Volume
df["Volume_Anomaly"] = np.where(df["Transaction Volume"] > df["Rolling_Volume_Mean"] + 3 * df["Rolling_Volume_STD"], 1, 0)

# Detecting Unusual Price Movements
df["Price_Anomaly"] = np.where(df["Transaction Price"] > df["Rolling_Price_Mean"] + 3 * df["Rolling_Price_STD"], 1, 0)

# Detecting Repeated Large Trades by the Same Entity
threshold = df["Transaction Volume"].quantile(0.95)
df["Repeated_Large_Trades"] = df.groupby("Trader_ID")["Transaction Volume"].transform(lambda x: 1 if x.mean() > threshold else 0)

# Detecting Suspiciously Clustered Transactions (Entities making identical trades)
df["Cluster_Anomaly"] = df.duplicated(subset=["Transaction Date", "Transaction Volume"], keep=False).astype(int)

# Detecting Insider Trading Driven by Announcements
df["Pre_Announcement_Trade"] = np.where(
    (df["Trade_Before_Announcement"] == 1) & (df["Volume_Anomaly"] == 1), 1, 0
)

# Combine All Anomalies
df["Final_Anomaly"] = df[[
    "Volume_Anomaly", "Price_Anomaly", "Repeated_Large_Trades", "Cluster_Anomaly", "Pre_Announcement_Trade"
]].sum(axis=1)
df["Final_Anomaly"] = df["Final_Anomaly"].apply(lambda x: 1 if x > 0 else 0)

# Explanation of Anomalies
def explain_anomaly(row):
    explanations = []
    if row["Volume_Anomaly"] == 1:
        explanations.append("Unusual trading volume")
    if row["Price_Anomaly"] == 1:
        explanations.append("Unusual price movement")
    if row["Repeated_Large_Trades"] == 1:
        explanations.append("Frequent large trades by same entity")
    if row["Cluster_Anomaly"] == 1:
        explanations.append("Clustered transactions by multiple entities")
    if row["Pre_Announcement_Trade"] == 1:
        explanations.append("High trading volume before announcement")
    return ", ".join(explanations)

df["Anomaly_Explanation"] = df.apply(explain_anomaly, axis=1)

# Create a detailed anomaly explanation table
anomaly_explanations = []

for idx, row in df[df["Final_Anomaly"] == 1].iterrows():
    explanation = {
        "Anomaly ID": f"ANOM-{idx}",  # Unique Anomaly ID
        "Transaction Date": row["Transaction Date"],
        "Trader ID": row["Trader_ID"] if "Trader_ID" in row else "Unknown",
        "Security Description": row["Security Description"] if "Security Description" in row else "Unknown",
        "Security ID (CUSIP)": row["Security ID"] if "Security ID" in row else "Unknown",
        "Transaction Type": row["Transaction Type"],
        "Actual Transaction Volume (Millions)": row["Transaction Volume"],
        "Normal Transaction Volume (Millions)": row["Rolling_Volume_Mean"],
        "Actual Transaction Price": row["Transaction Price"],
        "Normal Transaction Price": row["Rolling_Price_Mean"],
        "Total Amount Transferred (Millions)": row["Total Amount Transferred"],
        "Anomaly Type": row["Anomaly_Explanation"],
        "Deviation from Normal": "",
        "Linked Announcement": row["Event Type"] if "Event Type" in row else "Unknown",
        "Market Impact": row["Market Impact"] if "Market Impact" in row else "Unknown"
    }

    # Explain deviation from normal
    if row["Volume_Anomaly"] == 1:
        deviation = row["Transaction Volume"] / row["Rolling_Volume_Mean"]
        explanation["Deviation from Normal"] = f"Trade volume was {deviation:.2f}x higher than normal (Normal: {row['Rolling_Volume_Mean']:.2f}M)."

    if row["Price_Anomaly"] == 1:
        deviation = row["Transaction Price"] / row["Rolling_Price_Mean"]
        explanation["Deviation from Normal"] += f" Price was {deviation:.2f}x higher than rolling mean (Normal: {row['Rolling_Price_Mean']:.2f})."

    if row["Pre_Announcement_Trade"] == 1:
        explanation["Deviation from Normal"] += " Trade occurred just before a major announcement."

    anomaly_explanations.append(explanation)

# Convert to DataFrame
anomaly_report_df = pd.DataFrame(anomaly_explanations)

# Save report
anomaly_report_df.to_csv("anomaly_report.csv", index=False)
print("âœ… Detailed Anomaly Report Saved: anomaly_report.csv")

# Print report preview
print("\nðŸ“Š Anomaly Explanation Report Preview:")
print(anomaly_report_df.head())

# Visualization
plt.figure(figsize=(14, 7))
plt.scatter(df["Transaction Date"], df["Transaction Volume"], color="blue", label="Normal Transactions")
plt.scatter(df[df["Final_Anomaly"] == 1]["Transaction Date"], df[df["Final_Anomaly"] == 1]["Transaction Volume"], color="red", label="Anomalies")
plt.xlabel("Transaction Date")
plt.ylabel("Transaction Volume")
plt.title("ðŸ“‰ Insider Trading Anomalies in OMO Transactions")
plt.legend()
plt.show()
```
